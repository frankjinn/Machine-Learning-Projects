```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a range for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute norms for each point
def l0_5_norm(w):
    return np.sqrt(np.abs(w))

def l1_norm(w):
    return np.abs(w)

def l2_norm(w):
    return w**2

# Compute norms on the grid
Z_0_5 = l0_5_norm(X) + l0_5_norm(Y)
Z_1 = l1_norm(X) + l1_norm(Y)
Z_2 = np.sqrt(l2_norm(X) + l2_norm(Y))

print((Z_2).shape)

# Plot
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.contour(X, Y, Z_0_5, levels=[1, 2, 3, 4, 5])
plt.title('0.5 Norm')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 3, 2)
plt.contour(X, Y, Z_1, levels=[1, 2, 3, 4, 5])
plt.title('1 Norm')
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(1, 3, 3)
plt.contour(X, Y, Z_2, levels=[1, 2, 3, 4, 5])
plt.title('2 Norm')
plt.xlabel('x')
plt.ylabel('y')

# plt.tight_layout()
plt.show()
```
